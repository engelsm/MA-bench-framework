```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df1 = pd.read_csv("/home/mengelsl/MA-bench-framework/eigen/outputs/perf_20251220_193232/perf_results.csv")
# take the median 
df1 = df1.groupby(['cores', 'algorithm']).median().reset_index()
```

```python
#calculate speedup compared to 1 core
df1['speedup'] = df1.apply(lambda row: df1[(df1['cores'] == 1) & (df1['algorithm'] == row['algorithm'])]['real_time_s'].values[0] / row['real_time_s'], axis=1)
```

```python
sns.barplot(data=df1, x='cores', y='speedup', hue='algorithm')
```
