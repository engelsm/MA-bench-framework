```python
import numpy as np
import pandas as pd

def print_stats(df):
	print("Mean Time per Binary")
	t_per_binary= df.groupby(['Binary'])['Time'].mean()
	print(t_per_binary)
	print("Difference in Mean Time between spmv_DYNAMIC and spmv")
	t_per_binary_diff_pct = (t_per_binary['spmv_DYNAMIC'] - t_per_binary['spmv']) / t_per_binary['spmv'] * 100
	print(t_per_binary_diff_pct)
	print("\nStandard Deviation of Mean Times across Nodes")
	means = df.groupby(['Binary', 'Node'])['Time'].mean()
	std_per_binary = means.groupby('Binary').std()
	print(std_per_binary)
	print("Factor of Standard Deviation between spmv_DYNAMIC and spmv")
	std_per_binary_factor = std_per_binary['spmv_DYNAMIC'] / std_per_binary['spmv']
	print(std_per_binary_factor)

df1 = pd.read_csv("results.csv") #Cores:24 Matrix: 0-0_N1440352
df2 = pd.read_csv("results2.csv") #Cores:8 Matrix: 0-0_N432105

print("Results df1:")
print_stats(df1)
print("\nResults df2:")
print_stats(df2)
```
