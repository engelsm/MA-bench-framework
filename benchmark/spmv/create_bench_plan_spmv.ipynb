```python
import pandas as pd
import numpy as np

input_file = "itertest_spmv5.csv"
df = pd.read_csv(input_file)
# csv columns: Matrix,Cores,MemoryPolicy,Iterations,Runtime,Gflops

df["Iter_Target"] = ((df["Iterations"] / df["Runtime"]) * 5).round().astype(int)

plan_data = []
for _, row in df.iterrows():
    plan_data.append(row[["Matrix", "Cores", "Iter_Target"]].tolist())

plan_df = pd.DataFrame(plan_data, columns=["Matrix", "Cores", "Iter"])
plan_df.to_csv("bench_plan.csv", index=False)

print("Bench plan created and saved.")
```
