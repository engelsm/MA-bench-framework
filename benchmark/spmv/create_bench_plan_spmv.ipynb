```python
import pandas as pd
import numpy as np

input_file = "itertest_spmv.csv"
df = pd.read_csv(input_file)

df["Iter_Target"] = (df["Iterations"] / df["Runtime"]) * 5

plan_data = []
for _, row in df.iterrows():
    matrix = row["Matrix"]
    cores = int(row["Cores"])
    current_iter = int(row["Iter_Target"])

    if cores > 24:
        plan_data.append([matrix, cores, "localalloc", current_iter])
        plan_data.append([matrix, cores, "interleave", current_iter])
    else:
        plan_data.append([matrix, cores, "localalloc", current_iter])

plan_df = pd.DataFrame(plan_data, columns=["Matrix", "Cores", "MemoryPolicy", "Iter"])
plan_df.to_csv("bench_plan.csv", index=False)

print("Bench plan created and saved.")
```
