```python
import pandas as pd
import numpy as np

input_file = "itertest_krylov.csv"
df = pd.read_csv(input_file)

TARGET_TIME = 5.0


df["Time_per_Op"] = df["SpMV_Time"] / df["N_OPS"]
df["Target_Ops"] = TARGET_TIME / df["Time_per_Op"]

plan_data = []

for _, row in df.iterrows():
    algo = row["Algo"]
    matrix = row["Matrix"]
    cores = int(row["Cores"])
    
    scaling_factor = df["Target_Ops"].iloc[_] / row["N_OPS"]
    
    new_arg1 = int(max(1, row["Arg1"] * scaling_factor))
    
    arg2 = int(row["Arg2"])
    arg3 = int(row["Arg3"])

    policies = ["localalloc"]
    if cores > 24:
        policies.append("interleave")

    for policy in policies:
        plan_data.append([
            algo, 
            matrix, 
            cores, 
            policy, 
            new_arg1, 
            arg2, 
            arg3
        ])

plan_df = pd.DataFrame(plan_data, columns=["Algo", "Matrix", "Cores", "MemoryPolicy", "Arg1", "Arg2", "Arg3"])

plan_df.to_csv("bench_plan.csv", index=False)

print(f"Bench plan with {len(plan_df)} test runs created and saved to 'bench_plan.csv'.")
```
